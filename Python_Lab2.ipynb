```python
import pandas as pd
import numpy as np
# Load the Excel file
file_path = "LabSessionData.xlsx"
df = pd.read_excel(file_path, sheet_name="Purchase data")

# Display the dataframe
df
```

### Please load the data and segregate them into 2 matrices A & C (following the nomenclature of AX = C)

```python
df=df.iloc[:, :5]
df
# dropy = df.dropna()
# dropy
```

```python
df.columns
```

```python
A=df.iloc[:,1:4]
A
```

```python
C=df.iloc[:,4:]
C
```

### Array A and C to numpy Array

```python

A_arr=A.to_numpy()
A_arr
C_arr=C.to_numpy()
C_arr
```

### Rank of matrix A

```python
rank_A=np.linalg.matrix_rank(A_arr)
rank_A
```

### inverse of matrix A

```python
A_cost=np.linalg.pinv(A)
A_cost
```

### Matrix Multiplication

```python
cost_X=np.matmul(A_cost,C)
cost_X
```
